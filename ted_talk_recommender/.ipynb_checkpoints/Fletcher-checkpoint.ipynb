```python
df = pd.read_csv('TED_Talks.csv',nrows = 10000)
df.head(10)
```

```python
df.shape
```

```python
d = df.description
d = d.str.replace('[^\sa-zA-Z]|_', '')
t = df.transcript
t = t.str.replace('[^\sa-zA-Z]|_', '')
```

```python
d_sw = [] + list(esw)
t_sw = [] + list(esw)

dvectorizer = TfidfVectorizer(stop_words=d_sw,min_df=.03,max_df=.05)
tvectorizer = TfidfVectorizer(stop_words=t_sw,min_df=.01,max_df=.2)
```

```python
t.head()
```

```python
dtransformed = dvectorizer.fit_transform(d)
ttransformed = tvectorizer.fit_transform(t.values.astype(str))
```

```python
corpus = pd.DataFrame(ttransformed.toarray(), columns=tvectorizer.get_feature_names())
corpus.head()
```

```python
def display_topics(model, feature_names, no_top_words, topic_names=None):
    for ix, topic in enumerate(model.components_):
        if not topic_names or not topic_names[ix]:
            print("\nTopic ", ix)
        else:
            print("\nTopic: '",topic_names[ix],"'")
        print(", ".join([feature_names[i]
                        for i in topic.argsort()[:-no_top_words - 1:-1]]))
```

```python
def try_diff_num_topics(start=2,end=5, trans=None, vect=None):
    for n in range(start,end):
        print('')
        print(str(n)+' Topics')
        nmf = NMF(n)
        doc_topic = nmf.fit_transform(trans)
        ind = []
        for x in range(n): ind.append('component_'+str(x))
        topic_word = pd.DataFrame(nmf.components_.round(3), index = ind,columns = vect.get_feature_names())
        display_topics(nmf, vect.get_feature_names(), 9)
        
        print('')
        topid=[]
        for doc in doc_topic:
            topid.append(np.argmax(doc))
        print("Topics: "+str(n)+" "+str(Counter(topid)))
        
        print('')
        print("*********************************************")
    
```

```python
try_diff_num_topics(9,25, ttransformed, tvectorizer)
```

```python
def recommend_on_id(i=0,trans=None):
    d = pairwise_distances(trans,metric='cosine')
    return list(d[i].argsort()[:4])
```
